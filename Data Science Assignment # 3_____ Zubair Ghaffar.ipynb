### Creating a new Column number of cast in cast cell data

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('netflix_data.csv')
```

### Checking data is loaded already

```python
data.head(n=2)
```

### Checking null values in cast columns

```python
data.isnull().sum()
```

### Replacing null values with 0 to avoid any error in data processing while creating new column

```python
data['cast']=data['cast'].replace(np.nan,0)
```

### Checking again for null values in cast column

```python
data.isnull().sum()
```

```python
def cast_number(i):
    if i==0:
        return 0
    else:
        a=len(i.split(","))
        return a
```

```python
data['number_of_cast']= data["cast"].apply(cast_number)
```
