```python
import pandas as pd
```

```python
ds = pd .read_csv('Pokemon.csv')
```

```python
ds.head(n=10)
```

```python
avg_speed= ds["Speed"].mean()
avg_speed
```

```python
avg_hp = ds["HP"].mean()
avg_hp
```

```python
def set_speed_values(val):
    if val < avg_speed:
        return "Low Speed"
    elif val == avg_speed:
        return "Equal Speed"
    else:
        return "High Speed"
```

```python
def set_HP_values (value):
    if value < avg_hp:
        return "Low HP"
    elif value == avg_speed:
        return "Equal HP"
    else:
        return "High HP"
```

```python
ds["High_Low_Speed"]= ds["Speed"].apply(set_speed_values)
ds["High_Low_HP"]= ds["HP"].apply(set_HP_values)
```
